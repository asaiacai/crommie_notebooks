```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_excel('ET7_diffusion_set1_test.xlsx')
```

```python
data
```

```python
agg_data=data.drop_duplicates(subset = ['Gate','Voltage'],keep='last').pivot(columns='Gate',values='Voltage')
```

```python
columns = list(agg_data)
columns.pop(8)
```

```python
test = []
for col in columns:
    test.append(agg_data[col].dropna().values)
test = np.transpose(test)
```

```python
pd.DataFrame(test,columns=columns).to_excel('ET7_diffusion_set1_voltages.xlsx', index=False)
```
