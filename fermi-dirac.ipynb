```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rcParams.update({'font.size':20})
plt.figure(figsize=(10,3))
x = np.linspace(0,1,1000)
T=0.03
ef=.8
f = 1/(np.exp((x-ef)/T)+1)
plt.plot(x,f,linewidth=4,color='black')
plt.xlabel("$\epsilon$")
plt.ylabel("f($\epsilon$)")
plt.savefig('fermi-dirac-300K',bbox_inches='tight')
```

```python
mpl.rcParams.update({'font.size':20})
plt.figure(figsize=(10,3))
x = np.linspace(0,1,1000)
T=0.03/75
ef=.8
f = 1/(np.exp((x-ef)/T)+1)
plt.plot(x,f,linewidth=4, color='black')
plt.xlabel("$\epsilon$")
plt.ylabel("f($\epsilon$)")
plt.savefig('fermi-dirac-4K',bbox_inches='tight')
```
